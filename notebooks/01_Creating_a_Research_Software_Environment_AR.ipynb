<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<h2 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">إصدارات البرمجيات وقابلية إعادة الإنتاج</h2>
</div>

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">عند استخدام برمجيات بحثية في تحليل ما، فإن نتائج ذلك التحليل تعتمد على البرمجيات المستخدمة، بما في ذلك الإصدار المحدد منها. لإنشاء بحث قابل لإعادة الإنتاج باستخدام البرمجيات، نحتاج إلى تحديد جميع الاعتمادات البرمجية وإصداراتها بوضوح والتواصل بشأنها.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">تصبح المسألة أكثر تعقيدًا عندما نعترف بأن سلوك برمجياتنا يعتمد على برمجيات مفتوحة المصدر كتبها آخرون. كما قد تعتمد نتائج برنامج <span dir="ltr" style="unicode-bidi: isolate;">Python</span> على إصدار <span dir="ltr" style="unicode-bidi: isolate;">Python</span> نفسه. <strong>كيف يمكننا تتبع كل ذلك</strong>.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">في الدرس السابق تعلمنا استخدام <span dir="ltr" style="unicode-bidi: isolate;">pip</span> لتثبيت وإدارة حزم <span dir="ltr" style="unicode-bidi: isolate;">Python</span>. سابقًا كنا نثبت الحزم في دليل المستخدم بحيث تصبح متاحة في كل مرة نشغّل فيها <span dir="ltr" style="unicode-bidi: isolate;">Python</span>. يُشار إلى إصدار <span dir="ltr" style="unicode-bidi: isolate;">Python</span> وجميع الحزم المثبتة باسم بيئة <span dir="ltr" style="unicode-bidi: isolate;">Python</span>. كل مشروع جديد يمكنه الوصول إلى هذه الحزم نفسها.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><strong>قد يبدو ذلك مريحًا، لكنه يحمل بعض السلبيات.</strong> تخيل أننا سنبدأ مشروعًا جديدًا يستخدم ملفًا باسم <span dir="ltr" style="unicode-bidi: isolate;">example.py</span> يعتمد على <span dir="ltr" style="unicode-bidi: isolate;">NumPy</span> إصدار <span dir="ltr" style="unicode-bidi: isolate;">1.26.0</span>، بينما لدينا في نظامنا إصدار <span dir="ltr" style="unicode-bidi: isolate;">2.0.0</span>.</p><ul>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><span dir="ltr" style="unicode-bidi: isolate;">example.py</span> يعمل فقط مع الإصدار <span dir="ltr" style="unicode-bidi: isolate;">1.26.0</span> أو أقدم لأنه يستخدم <span dir="ltr" style="unicode-bidi: isolate;">np.infty</span> الذي تغير إلى <span dir="ltr" style="unicode-bidi: isolate;">np.inf</span> في الإصدار <span dir="ltr" style="unicode-bidi: isolate;">2.0.0</span>.</li>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;">مشروعنا الجديد يحتاج إلى <span dir="ltr" style="unicode-bidi: isolate;">NumPy</span> إصدار <span dir="ltr" style="unicode-bidi: isolate;">2.0.0</span> لدعم ميزة جديدة. إذا خفّضنا الإصدار فلن يعمل مشروعنا الجديد.</li>
</ul><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">يوضح هذا المثال المشكلات التي قد تظهر عند تشغيل مشاريع متعددة ضمن البيئة نفسها.</p>
</div>

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<h3 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">إدارة الإصدارات البرمجية</h3><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">يوجد طريقتان شائعتان للإشارة إلى إصدار البرمجيات.</p><ul>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;">إدارة إصدارات التقويم حيث يعكس الرقم تاريخ الإصدار بصيغة <span dir="ltr" style="unicode-bidi: isolate;">YYYY.MM</span>.</li>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;">إدارة الإصدارات الدلالية حيث يوضح الرقم مدى اختلاف الإصدار الجديد عن السابق.</li>
</ul><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><a href="https://semver.org/#semantic-versioning-200" dir="ltr">إدارة الإصدارات الدلالية</a> هي الأكثر شيوعًا، وتتكون من ثلاثة أجزاء مفصولة بنقاط.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><span dir="ltr" style="unicode-bidi: isolate;">MAJOR.MINOR.PATCH</span></p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">على سبيل المثال، يستخدم <span dir="ltr" style="unicode-bidi: isolate;">NumPy</span> هذا النظام في إصدارات مثل <span dir="ltr" style="unicode-bidi: isolate;">2.0.0</span> و <span dir="ltr" style="unicode-bidi: isolate;">1.26.0</span>.</p><ul>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;">الجزء <span dir="ltr" style="unicode-bidi: isolate;">MAJOR</span> يشير إلى تغييرات قد تكسر التوافق مع الإصدارات السابقة.</li>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;">الجزء <span dir="ltr" style="unicode-bidi: isolate;">MINOR</span> يشير إلى ميزات جديدة متوافقة مع الإصدارات السابقة.</li>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;">الجزء <span dir="ltr" style="unicode-bidi: isolate;">PATCH</span> يشير إلى إصلاحات أخطاء دون كسر التوافق.</li>
</ul>
</div>

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<h3 style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><span dir="ltr" style="unicode-bidi: isolate;">Python</span> بيئات افتراضية</h3><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">الحل لمشكلة تعارض الإصدارات هو إنشاء بيئات منفصلة لكل مشروع.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">أداة <span dir="ltr" style="unicode-bidi: isolate;">virtualenv</span> تسمح بإنشاء بيئات افتراضية مستقلة يمكن أن تحتوي على إصدارات مختلفة من <span dir="ltr" style="unicode-bidi: isolate;">Python</span> والحزم.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><img src="./assets/M3_venv.png" alt="virtualenv diagram" dir="ltr"></p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">يوضح الشكل أن كل بيئة افتراضية يمكن أن تحتوي على إصدارات مختلفة من الحزم مثل <span dir="ltr" style="unicode-bidi: isolate;">numpy</span>.</p>
</div>

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<h3 style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><span dir="ltr" style="unicode-bidi: isolate;">Python</span> بيئة افتراضية إنشاء</h3><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">سنستخدم سطر الأوامر لإنشاء بيئة افتراضية داخل مجلد المشروع باسم <span dir="ltr" style="unicode-bidi: isolate;">venv</span>.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><span dir="ltr" style="unicode-bidi: isolate;">mkdir venv</span></p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">ثم ننشئ البيئة باسم <span dir="ltr" style="unicode-bidi: isolate;">h2o</span>.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><span dir="ltr" style="unicode-bidi: isolate;">virtualenv venv/h2o</span></p>
</div>

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">لبدء العمل داخل البيئة الافتراضية نفعّلها باستخدام الأمر التالي.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><span dir="ltr" style="unicode-bidi: isolate;">source venv/h2o/bin/activate</span></p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">سيظهر اسم البيئة مثل <span dir="ltr" style="unicode-bidi: isolate;">(h2o)</span> في بداية سطر الأوامر.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">للخروج من البيئة نستخدم الأمر.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><span dir="ltr" style="unicode-bidi: isolate;">deactivate</span></p>
</div>

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<h3 style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><span dir="ltr" style="unicode-bidi: isolate;">pip</span> استخدام في <span dir="ltr" style="unicode-bidi: isolate;">Python</span> بيئة افتراضية</h3><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">يمكن تثبيت الحزم داخل البيئة باستخدام <span dir="ltr" style="unicode-bidi: isolate;">pip</span>.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><span dir="ltr" style="unicode-bidi: isolate;">pip install numpy==1.26.0</span></p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">وإزالة الحزم باستخدام الأمر.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><span dir="ltr" style="unicode-bidi: isolate;">pip uninstall numpy</span></p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">هذه التغييرات تؤثر فقط على البيئة الحالية ولا تؤثر على البيئات الأخرى أو تثبيت <span dir="ltr" style="unicode-bidi: isolate;">Python</span> الافتراضي.</p>
</div>

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<h3 style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><span dir="ltr" style="unicode-bidi: isolate;">Jupyter Notebook</span> استخدام في <span dir="ltr" style="unicode-bidi: isolate;">Python</span> بيئة افتراضية</h3><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">عند فتح دفتر <span dir="ltr" style="unicode-bidi: isolate;">Jupyter Notebook</span> يمكن اختيار النواة <span dir="ltr" style="unicode-bidi: isolate;">kernel</span> التي تنفذ الشيفرة.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><img src="./assets/M3_Jupyter_kernel.png" alt="Jupyter kernel" dir="ltr"></p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">قد لا يتعرف <span dir="ltr" style="unicode-bidi: isolate;">Jupyter Notebook</span> تلقائيًا على بيئتنا الافتراضية.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">نثبت الحزمة <span dir="ltr" style="unicode-bidi: isolate;">ipykernel</span> داخل البيئة.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><span dir="ltr" style="unicode-bidi: isolate;">pip install ipykernel</span></p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">ثم نسجل النواة الجديدة باسم <span dir="ltr" style="unicode-bidi: isolate;">OpenScience</span>.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><span dir="ltr" style="unicode-bidi: isolate;">python -m ipykernel install --user --name=OpenScience</span></p><h3 style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><span dir="ltr" style="unicode-bidi: isolate;">Jupyter Notebook</span> إعادة تشغيل</h3><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">بعد إعادة التشغيل نختار النواة المناسبة من القائمة.</p><p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><strong>أفضل ممارسة</strong> هي أن يتطابق اسم البيئة الافتراضية مع اسم النواة.</p>
